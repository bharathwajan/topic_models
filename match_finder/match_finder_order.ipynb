```python
import pandas as pd

def accuracy(gntm_path,baysian_path,n_largest,match_threshold):
    bay=pd.read_csv(f"C:/Users/rbw19/OneDrive/Desktop/nb_gnn_comp/distribition_finder/{baysian_path}")  #
    gnt=pd.read_csv(f"C:/Users/rbw19/OneDrive/Desktop/nb_gnn_comp/distribition_finder/{gntm_path}")
    print("read sucessfully")
    bay.drop(["Unnamed: 0"],axis=1,inplace=True) 
    gnt.drop(["Unnamed: 0"],axis=1,inplace=True) 
    bay,gnt=bay.transpose(),gnt.transpose()
    bay_top_3,gnt_top_3=[],[]
    for col in range(bay.shape[1]):
        try:
            bay_top_3.append(bay.nlargest(n_largest,col).index) #parameter passs
        except:
            print("inside except")
            pass

    for col in range(gnt.shape[1]):
        try:
            gnt_top_3.append(gnt.nlargest(n_largest,col).index)
        except:
            print("inside except")
            pass
    
    final_match=[]
    for gnt_data,bay_data in zip(gnt_top_3,bay_top_3):
        # print('inside',gnt_data,bay_data )
        matches=[]
        for gnt_,bay_ in zip(gnt_data,bay_data):
            if gnt_==bay_:
                # print(True)
                matches.append(True)
            else:
                matches.append(False)

        selection_threshold=(matches.count(True)/len(matches))*100
        if selection_threshold>=match_threshold:
            final_match.append(True)
        else:
            final_match.append(False)
    print(len(final_match),gnt.shape,bay.shape)     


    accuracy=(final_match.count(True)/len(final_match))*100
    print("Accuracy is :",accuracy)

    # decision=str(input("Do you want to save the o/p(yes/no)?"))
    decision='no' # for temporaryly
    if decision=="Yes" or decision=="yes" or decision=="YES":
        bay_col_names=[]
        gnt_col_names=[]
        for i in range(1,n_largest+1):
            bay_col_names.append(f'bay_top{str(i)}')
            gnt_col_names.append(f'gnt_top{str(i)}')

        bay_top3_df=pd.DataFrame(bay_top_3,columns=bay_col_names)
        gnt_top3_df=pd.DataFrame(gnt_top_3,columns=gnt_col_names)
        result=pd.concat([bay_top3_df,gnt_top3_df,pd.DataFrame(final_match,columns=["match"])],axis=1)
        file_name=str(input("enter file name :"))

        result.to_csv(f"{file_name}.csv")



# topics=str(input("enter no. of topics :"))
# n_largest=int(input("enter n_largest : "))
# match_threshold=int(input("Enter the threshold value : "))  #80 percentage 

# gntm_path=f"gntm_topic_{topics}.csv"
# baysian_path=f"baysian_topic_{topics}.csv"
# accuracy(gntm_path,baysian_path,n_largest,match_threshold)

import math
topics=[5,7,10,12,20]
def ceilorfloor(val):
    frac, whole = math.modf(val)
    if frac>0.5:
        return math.ceil(val)
    elif frac<0.5:
        return math.floor(val)


for topic in topics:
    sixty=  ceilorfloor(0.6*topic)
    eighty= ceilorfloor(0.8*topic)
    hundred=ceilorfloor(topic)
    # print('topic',topic)
    n_largest=[sixty,eighty,hundred]
    for val in n_largest:
        for percentage in [60,80,100]:
            print('\n No of topics :',topic,'\n Nlargest :',val,'\n Match threshold :',percentage)
            gntm_path=f"gntm_topic_{topic}.csv"
            baysian_path=f"baysian_topic_{topic}.csv"
            accuracy(gntm_path,baysian_path,val,percentage)
```
