```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# without_order=pd.read_csv("without_order.csv")
without_order=pd.read_csv("news_20_plsa.csv")
# with_order=pd.read_csv("with_order.csv")
# without_order.columns
```

```python
topics=without_order["No of topics"].unique() #--> x axis values
topics
```

# 80% n largest

```python
sixty_without_order=without_order.loc[(without_order["Top n largest percentage"]=="80%") & (without_order["Match threshold"]==60)]["Accuracy"].values
eighty_without_order=without_order.loc[(without_order["Top n largest percentage"]=="80%") & (without_order["Match threshold"]==80)]["Accuracy"].values
hundred_without_order=without_order.loc[(without_order["Top n largest percentage"]=="80%") & (without_order["Match threshold"]==100)]["Accuracy"].values

# sixty_with_order=with_order.loc[(with_order["Top n largest percentage"]=="80%") & (with_order["Match threshold"]==60)]["Accuracy"].values
# eighty_with_order=with_order.loc[(with_order["Top n largest percentage"]=="80%") & (with_order["Match threshold"]==80)]["Accuracy"].values
# hundred_with_order=with_order.loc[(with_order["Top n largest percentage"]=="80%") & (with_order["Match threshold"]==100)]["Accuracy"].values

sixty_without_order=sixty_without_order*100
eighty_without_order=eighty_without_order*100
hundred_without_order=hundred_without_order*100
# eighty_without_order*100
```

```python
plt.plot(topics,sixty_without_order,color='b',linewidth=2,label='60% match_wo',marker='.')
plt.plot(topics,eighty_without_order,color='r',linewidth=2,label='80% match_wo',marker='.')
plt.plot(topics,hundred_without_order,color='g',linewidth=2,label='100% match_wo',marker='.')

# plt.plot(topics,sixty_with_order,color='b',linewidth=2,linestyle='--',label='60% match_o',marker='.')
# plt.plot(topics,eighty_with_order,color='r',linewidth=2,linestyle='--',label='80% match_o',marker='.')
# plt.plot(topics,hundred_with_order,color='g',linewidth=2,linestyle='--',label='100% match_o',marker='.')

plt.title('80 % top n largest VS ACCURACY',fontsize=18,loc='left')
plt.xlabel("NO OF TOPICS")  
plt.ylabel("ACCURACY")
#plt.xlim(0,25)
#plt.ylim(50,110)

plt.legend()
plt .grid()
plt.show()
```
